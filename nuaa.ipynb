```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model


from data.DataLoader import NUAADataLoader
from training.trainModel import trainModel, benchmarkModel
```

```python
data_source =  "../input/nuaaraw/NUAA/"
saved_models = "trained_models/"
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
dtLoader = NUAADataLoader(data_source)
dtLoader.loadData()
```

```python
train_raw = dtLoader.getTrainSet('raw')
valid_raw = dtLoader.getValidSet('raw')

train_norm = dtLoader.getTrainSet('normalized')
valid_norm = dtLoader.getValidSet('normalized')

train_det = dtLoader.getTrainSet('detected')
valid_det = dtLoader.getValidSet('detected')
```

```python
model_raw, h_raw = trainModel(train_raw, valid=valid_raw, ds_name='raw', epochs=10, theta=0.4)
```

```python
model_raw = load_model(saved_models+'model_raw')
```

```python
s_raw = benchmarkModel(model_raw, valid_raw)
```

```python
model_normalized, h_normalized = trainModel(train_norm, valid=valid_norm, ds_name='normalized', epochs=10, theta=0.7)
```

```python
model_normalized = load_model(saved_models+'model_normalized')
```

```python
s_normalized = benchmarkModel(model_normalized, valid_norm)
```

```python
model_detected, h_detected = trainModel(train_det, valid=valid_det, ds_name='detected', epochs=10, theta=0.7)
```

```python
model_detected = load_model(saved_models+'model_detected')
```

```python
s_detected = benchmarkModel(model_detected, valid_det)
```
